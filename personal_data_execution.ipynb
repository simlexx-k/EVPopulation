```python
#Loading Data

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#from scipy.stats import gaussian_kde

%matplotlib inline
```

```python
pdata = pd.read_csv('data/personal_data.csv')

print('Data Overview:')
pdata.head()
```

```python
print('Basic Statistical Information:')
pdata.describe()
```

```python
print('Brief Information:')
pdata.info()
```

```python
#Define datapoints

x = pdata['gender      '].isin(['Female', 'Male'])
y = pdata['family_size']
# Plot the data
plt.plot(x, y, label='Gender vs Family Size')
plt.title('Gender vs Family Size Data Visualization')
plt.xlabel('X-axis - ' + 'Gender')
plt.ylabel('Y-axis - ' + 'Family Size')
plt.legend()
plt.show()
```

```python
#Sort by balance
asort = pdata.sort_values(['balance   '], ascending= False)
print('Data Sorted according to Balances (Descending Order):')
asort
```

```python
bsort = pdata.sort_values(['balance   '], ascending= True)
print('Data Sorted according to Balances (Ascending Order):')
bsort
```

```python
#Bar graph plot of vehicles by model
print('Number of specific vehicle models:')
datafig = pdata['vehicle       '].value_counts()
veh_counts = pdata['vehicle       '].value_counts().plot(kind='bar', figsize=[15,8], xlabel='Vehicle', ylabel='Number')
veh_counts
datafig
```
